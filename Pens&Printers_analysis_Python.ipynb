Pens and Printers was founded in 1984 and provides high quality office products to large organizations. We are a trusted provider of everything from pens and notebooks to desk chairs and monitors. We don’t produce our own products but sell those made by other companies.

We have built long lasting relationships with our customers and they trust us to provide them with the best products for them. ### New Product Sales Methods

Six weeks ago we launched a new line of office stationery. Despite the world becoming increasingly digital, there is still demand for notebooks, pens and sticky notes.

Our focus has been on selling products to enable our customers to be more creative, focused on tools for brainstorming. We have tested three different sales strategies for this, targeted email and phone calls, as well as combining the two:

- Email: Customers in this group received an email when the product line was launched, and a further email three weeks later. This required very little work for the team.
- Call: Customers in this group were called by a member of the sales team. On average members of the team were on the phone for around thirty minutes per customer.
- Email and call: Customers in this group were first sent the product information email, then called a week later by the sales team to talk about their needs and how this new product may support their work. The email required little work from the team, the call was around ten minutes per customer.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats

sales = pd.read_csv('product_sales.csv')
print(sales.info())

print(sales.isna().sum())

print(sales['week'].describe())
print(sales['sales_method'].value_counts()) # text not same
print(sales['customer_id'].head())
print(sales['nb_sold'].describe()) 
print(sales['revenue'].describe()) # 1074 NAs
print(sales['years_as_customer'].describe()) # outlier
print("Years above 40:", sales[sales['years_as_customer'] > 40].shape[0])
print(sales['nb_site_visits'].describe())
print(sales['state'].value_counts())
print("Number of states is:", sales['state'].nunique())

```

```python
#Make a copy of the sales dataframe
df = sales.copy()

#Uniform text
df['sales_method'] = df['sales_method'].str.title()
replace_map = {'Em + Call':'Email + Call'}
df['sales_method'].replace(replace_map, inplace=True)
print(df['sales_method'].value_counts())

#View outlier. Company started 1984
sns.scatterplot(x='years_as_customer', y='revenue', data=df)
plt.show()
max_year_change = 40
#drop values > 40 since I am unsure how to proceed with them otherwise.
df.drop(df[df['years_as_customer'] > 40].index, inplace=True)

#Impute NAs with mean since values are floats
#Checking if revenue correlates with any of the other values

#look also at sales method variables
#convert to numerical
sales_method_map = {'Email': 0,
                    'Call': 1,
                    'Email + Call': 2}
df1 = df.copy()
df1['sales_method'] = df1['sales_method'].replace(sales_method_map)
sns.heatmap(df1.corr(), annot=True) #High correlation with nb_sold at 0.7
#High correlation between week and nb_sold also observed of 0.8

#Imputing revenue NAs with mean revenue by number of products sold
print("Old revenue:", df['revenue'].mean())
df['revenue'] = df['revenue'].fillna(df.groupby('nb_sold')['revenue'].transform('mean'))
print("New revenue:", df['revenue'].mean())
 

duplicates = df.duplicated()
print(df[duplicates]) # no duplicates

#Check new cleaned dataframe
print(df.info())
```

```python
#Count of each sales method
sales_order= ['Email','Call','Email + Call']
g = sns.catplot(x='sales_method', data=df, kind='count', order=sales_order)
g.set_axis_labels("Sales Method", "Count/Number of Customers")
plt.savefig("Count_sales_method.png",dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Plotting versus revenue

#revenue spread
g = sns.boxplot(y='revenue', data=df) 
g.set(ylabel="Revenue")
plt.savefig("Revenue_spread.png", dpi=300, bbox_inches="tight")
plt.show()
print(df['revenue'].describe())

#Calculate total revenue for each sales method and create a new DataFrame
total_revenue_by_method = df.groupby('sales_method')['revenue'].sum().reset_index()
total_revenue_by_method.rename(columns={'revenue': 'total_revenue'}, inplace=True)
print(total_revenue_by_method)

#Plot average revenue per sales method
g = sns.catplot(x='sales_method', y='revenue', data=df, kind='box', order=sales_order) 
g.set_axis_labels("Sales Method", "Revenue")
plt.savefig("Revenue_per_sales_method.png", dpi=300, bbox_inches="tight")
plt.show()

#Plot total revenue per sales method
h = sns.catplot(x='sales_method', y='total_revenue', data=total_revenue_by_method, kind='bar', order=sales_order)
h.set_axis_labels("Sales Method", "Total Revenue")
plt.savefig("Total_Revenue_per_sales_method.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Plotting versus revenue and week
g = sns.catplot(x='week', y='revenue', data=df, kind='bar', hue='sales_method', hue_order=sales_order)
g.set_axis_labels("Week", "Revenue")
plt.savefig("Revenue_sales_method_by_week.png",dpi=300, bbox_inches="tight")
plt.show()

#Calculate total revenue per week and sales method and create a new DataFrame
total_revenue_sm = df.groupby(['week','sales_method'])['revenue'].sum().reset_index()
total_revenue_sm.rename(columns={'revenue': 'total_revenue'}, inplace=True)

#Plotting versus total revenue and week
g = sns.catplot(x='week', y='total_revenue', data=total_revenue_sm, kind='bar', hue='sales_method', hue_order=sales_order)
g.set_axis_labels("Week", "Total Revenue")
plt.savefig("Total_Revenue_sales_method_by_week.png",dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Plotting versus nb_sold

g = sns.catplot(x='sales_method', y='nb_sold', data=df, kind='bar', order=sales_order)
g.set_axis_labels("Sales Method", "Number of Products Sold")
plt.savefig("Nb_Sold_per_Sales_method.png",dpi=300, bbox_inches="tight")
plt.show()

```

```python
#Plotting revenue per week

g = sns.lineplot(x='week', y='revenue', data=df)
g.set(xlabel="Week", ylabel="Revenue")
plt.savefig("Revenue_Over_6_weeks.png",dpi=300, bbox_inches="tight")
plt.show()

#Calculate total revenue per week and create a new DataFrame
total_revenue_by_week = df.groupby('week')['revenue'].sum().reset_index()
total_revenue_by_week.rename(columns={'revenue': 'total_revenue'}, inplace=True)

#Plot total revenue per sales method
g = sns.lineplot(x='week', y='total_revenue', data=total_revenue_by_week)
g.set(xlabel="Week", ylabel="Total Revenue")
plt.savefig("Total_Revenue_by_week.png", dpi=300, bbox_inches="tight")
plt.show()

print(total_revenue_by_week.set_index('week'))
print("mean is:", total_revenue_by_week.mean())
print("std is:", total_revenue_by_week.std())

#Plot count of customers per week
g = sns.catplot(x='week', data=df, kind='count')
g.set_axis_labels("Week", "Count/Number of Customers")
plt.savefig("Count_sales_per_week.png",dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Plotting revenue per strategy count

df['method_count'] = df.groupby('sales_method')['sales_method'].transform('count')

# Divide total revenue by the count of each sales_method type
df['rev_per_method_count'] = df['revenue'].sum() / df['method_count']

g = sns.catplot(x='sales_method', y='rev_per_method_count', data=df, kind='bar', order=sales_order)  
g.set_axis_labels("Sales Method", "Total Revenue per Customer")
plt.savefig("Total_revenue_per_sales_method.png",dpi=300, bbox_inches="tight")
plt.show()

Mean = df.groupby('sales_method')['rev_per_method_count'].mean()
print(Mean)
Fold_over_email = Mean["Email + Call"]/Mean["Email"]
Fold_over_call= Mean["Email + Call"]/Mean["Call"]
print("Email + Call is ", Fold_over_email, "times higher than email alone")
print("Email + Call is ", Fold_over_call, "times higher than call alone")
```

```python
# years as customer vs revenue

#Calculate total revenue vs years as customer and create a new DataFrame
total_revenue_by_years = df.groupby('years_as_customer')['revenue'].sum().reset_index()
total_revenue_by_years.rename(columns={'revenue': 'total_revenue'}, inplace=True)
print(total_revenue_by_years)

#Plot total revenue per sales method
g = sns.lineplot(x='years_as_customer', y='total_revenue', data=total_revenue_by_years)
g.set(xlabel="Years as customer", ylabel="Total Revenue")
plt.savefig("Total_Revenue_years.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# nb_site_visits vs revenue

#Calculate total revenue vs nb sites visits and sales method and create a new DataFrame
total_revenue_by_nb_visit = df.groupby(['nb_site_visits','sales_method'])['revenue'].sum().reset_index()
total_revenue_by_nb_visit.rename(columns={'revenue': 'total_revenue'}, inplace=True)

g = sns.lineplot(x='nb_site_visits', y='total_revenue', hue='sales_method', hue_order=sales_order, data=total_revenue_by_nb_visit) 
g.set(xlabel="Number of Site Visits", ylabel="Total Revenue")
plt.savefig("Total_Revenue_nb_visits.png", dpi=300, bbox_inches="tight")

plt.show()
```
